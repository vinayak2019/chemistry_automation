## Using *os*

```python
import os
```

```python
os.system("/content/xtb-dist/bin/xtb --versio")
```

Return value 0 indicates the process is successfully completed. Other return values indicate failure. Below is a simple script to check for successful completion.

```python
return_code = os.system("/content/xtb-dist/bin/xtb --version")
if return_code == 0:
  print("Success")
else:
  print("Failure")
```
